```python

class BaseAgent:
    COLORS = {
        'profiler': ('🔍', '#e94560'),
        'cleaner':  ('🧹', '#48cae4'),
        'eda':      ('📊', '#a8dadc'),
        'ml':       ('🤖', '#ffb74d'),
        'reporter': ('📝', '#81c784'),
        'default':  ('⚙️', '#ce93d8'),
    }

    def __init__(self, name: str, role: str):
        self.name  = name
        self.role  = role
        self.logs  = []
        self.state = {}
        self._icon, self._color = self.COLORS.get(role, self.COLORS['default'])

    def log(self, message: str, level: str = 'INFO'):
        ts = datetime.now().strftime('%H:%M:%S')
        self.logs.append(f"[{ts}] [{level}] {message}")

    def banner(self, title: str = None):
        label = title or f"{self._icon} {self.name}"
        display(HTML(f"""
        <div style='background:linear-gradient(135deg,{self._color}22,{self._color}11);
                    border-left:4px solid {self._color};
                    padding:14px 20px; border-radius:8px; margin:12px 0;
                    font-family:monospace;'>
          <span style='color:{self._color}; font-size:1.2em; font-weight:bold;'>{label}</span>
          <span style='color:#8b949e; font-size:0.85em; margin-left:12px;'>
            activated at {datetime.now().strftime('%H:%M:%S')}
          </span>
        </div>"""))

    def status(self, msg: str, ok: bool = True):
        icon  = '✅' if ok else '⚠️'
        color = '#81c784' if ok else '#ffb74d'
        display(HTML(
            f"<p style='color:{color}; font-family:monospace; margin:4px 0;'>{icon} {msg}</p>"
        ))

    def run(self, data, context: dict = None):
        raise NotImplementedError



class MultiAgentOrchestrator:
    def __init__(self):
        self.agents   = {}
        self.context  = {}
        self.pipeline = []
        self._history = []

    def register(self, agent: BaseAgent):
        self.agents[agent.name] = agent
        self.pipeline.append(agent.name)
        print(f"  {agent._icon} Registered: {agent.name}")

    def run_pipeline(self, data):
        display(HTML("""
        <div style='background:linear-gradient(135deg,#1a1a2e,#0f3460);
                    padding:25px; border-radius:15px; text-align:center; margin:15px 0;'>
          <h2 style='color:#e94560; margin:0; font-size:1.8em;'> PIPELINE EXECUTION STARTED</h2>
          <p style='color:#8b949e; margin:8px 0 0;'>Launching all agents in sequence…</p>
        </div>"""))

        current = data
        for name in self.pipeline:
            agent   = self.agents[name]
            t0      = time.time()
            result  = agent.run(current, self.context)
            elapsed = time.time() - t0
            if result is not None:
                current = result
            self.context[name] = agent.state
            self._history.append({'agent': name, 'elapsed': round(elapsed, 2)})

        self._summary()
        return current

    def _summary(self):
        rows = "".join(
            f"<tr><td style='padding:8px 15px; color:#48cae4;'>{h['agent']}</td>"
            f"<td style='padding:8px 15px; color:#81c784;'>{h['elapsed']}s</td></tr>"
            for h in self._history
        )
        display(HTML(f"""
        <div style='background:#161b22; padding:20px; border-radius:12px;
                    margin-top:20px; border:1px solid #30363d;'>
          <h3 style='color:#e94560; margin:0 0 12px;'>📋 Pipeline Summary</h3>
          <table style='border-collapse:collapse; width:100%;'>
            <thead><tr style='background:#21262d;'>
              <th style='padding:8px 15px; color:#c9d1d9; text-align:left;'>Agent</th>
              <th style='padding:8px 15px; color:#c9d1d9; text-align:left;'>Time</th>
            </tr></thead>
            <tbody>{rows}</tbody>
          </table>
        </div>"""))

print(" BaseAgent & Orchestrator defined.")
```

```python
class DataProfilingAgent(BaseAgent):

    def __init__(self):
        super().__init__('DataProfilingAgent', 'profiler')

    def run(self, df: pd.DataFrame, context: dict = None) -> pd.DataFrame:
        self.banner("🔍 Data Profiling Agent")
        self.log('Profiling started')

        n_rows, n_cols = df.shape
        missing_total  = df.isnull().sum().sum()
        dup_rows       = df.duplicated().sum()
        mem_mb         = df.memory_usage(deep=True).sum() / 1024**2

        display(HTML(f"""
        <div style='display:flex; gap:12px; flex-wrap:wrap; margin:10px 0;'>
          {self._kpi('Rows',    f'{n_rows:,}',       '#e94560')}
          {self._kpi('Columns', f'{n_cols:,}',       '#48cae4')}
          {self._kpi('Missing', f'{missing_total:,}','#ffb74d')}
          {self._kpi('Dupes',   f'{dup_rows:,}',     '#ce93d8')}
          {self._kpi('Memory',  f'{mem_mb:.2f} MB',  '#81c784')}
        </div>"""))

        profile = self._column_profile(df)
        self._render_profile_table(profile)

        num_cols = df.select_dtypes(include=np.number).columns.tolist()
        if len(num_cols) >= 2:
            self._plot_correlation(df[num_cols])

        self._plot_missing(df)

        self.state = {
            'shape':         (n_rows, n_cols),
            'missing_total': int(missing_total),
            'dup_rows':      int(dup_rows),
            'num_cols':      num_cols,
            'cat_cols':      df.select_dtypes(include='object').columns.tolist(),
            'profile':       profile,
        }
        self.status(f"Profiling complete — {n_cols} columns analysed.")
        return df

    def _kpi(self, label, value, color):
        return (
            f"<div style='background:{color}22; border:1px solid {color};"
            f"border-radius:10px; padding:12px 20px; text-align:center; min-width:110px;'>"
            f"<div style='color:{color}; font-size:1.4em; font-weight:bold;'>{value}</div>"
            f"<div style='color:#8b949e; font-size:0.8em;'>{label}</div></div>"
        )

    def _column_profile(self, df):
        rows = []
        for col in df.columns:
            s        = df[col]
            null_pct = s.isnull().mean() * 100
            unique   = s.nunique()
            dtype    = str(s.dtype)
            if pd.api.types.is_numeric_dtype(s):
                extra = f"mean={s.mean():.2f}, std={s.std():.2f}, skew={s.skew():.2f}"
            else:
                top   = s.value_counts().index[0] if unique else 'N/A'
                extra = f"top='{top}'"
            rows.append({'Column': col, 'Dtype': dtype,
                         'Null%': f"{null_pct:.1f}%", 'Unique': unique, 'Stats': extra})
        return rows

    def _render_profile_table(self, profile):
        header = "<tr>" + "".join(
            f"<th style='padding:8px 12px; color:#c9d1d9; background:#21262d; text-align:left;'>{k}</th>"
            for k in profile[0].keys()) + "</tr>"
        body = ""
        for i, row in enumerate(profile):
            bg    = '#161b22' if i % 2 == 0 else '#0d1117'
            body += f"<tr style='background:{bg};'>" + "".join(
                f"<td style='padding:7px 12px; color:#8b949e; font-family:monospace;'>{v}</td>"
                for v in row.values()) + "</tr>"
        display(HTML(f"""
        <div style='overflow-x:auto; margin:10px 0;'>
        <table style='border-collapse:collapse; width:100%; font-size:0.9em;'>
          <thead>{header}</thead><tbody>{body}</tbody>
        </table></div>"""))

    def _plot_correlation(self, df_num):
        corr = df_num.corr()
        fig, ax = plt.subplots(figsize=(max(6, len(corr)*0.9), max(5, len(corr)*0.8)))
        mask = np.triu(np.ones_like(corr, dtype=bool))
        sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
                    center=0, ax=ax, linewidths=0.5, annot_kws={'size': 9, 'color': 'white'})
        ax.set_title('📊 Correlation Matrix', color='#e94560', fontsize=14, pad=12)
        plt.tight_layout(); plt.show()

    def _plot_missing(self, df):
        missing = df.isnull().mean().sort_values(ascending=False)
        missing = missing[missing > 0]
        if missing.empty:
            self.status("No missing values detected — clean dataset!"); return
        fig, ax = plt.subplots(figsize=(10, max(3, len(missing)*0.5)))
        colors  = [PALETTE[i % len(PALETTE)] for i in range(len(missing))]
        missing.plot.barh(ax=ax, color=colors, edgecolor='none')
        ax.set_xlabel('Missing Fraction')
        ax.set_title('🔴 Missing Values by Column', color='#e94560', fontsize=13)
        ax.axvline(0.3, color='#ffb74d', linestyle='--', alpha=0.6, label='30% threshold')
        ax.legend()
        plt.tight_layout(); plt.show()

print(" DataProfilingAgent defined.")
```

```python
class DataCleaningAgent(BaseAgent):

    def __init__(self, outlier_z: float = 3.0, drop_thresh: float = 0.6):
        super().__init__('DataCleaningAgent', 'cleaner')
        self.outlier_z   = outlier_z
        self.drop_thresh = drop_thresh

    def run(self, df: pd.DataFrame, context: dict = None) -> pd.DataFrame:
        self.banner(" Data Cleaning Agent")
        self.log('Cleaning started')
        steps = []
        df    = df.copy()

        high_miss = df.columns[df.isnull().mean() > self.drop_thresh].tolist()
        if high_miss:
            df.drop(columns=high_miss, inplace=True)
            steps.append(f"Dropped {len(high_miss)} high-missing cols: {high_miss}")

        n_before = len(df)
        df.drop_duplicates(inplace=True)
        steps.append(f"Removed {n_before - len(df)} duplicate rows")

        num_cols = df.select_dtypes(include=np.number).columns.tolist()
        if num_cols:
            df[num_cols] = SimpleImputer(strategy='median').fit_transform(df[num_cols])
            steps.append(f"Imputed {len(num_cols)} numerical cols with median")

        cat_cols = df.select_dtypes(include='object').columns.tolist()
        if cat_cols:
            df[cat_cols] = SimpleImputer(strategy='most_frequent').fit_transform(df[cat_cols])
            steps.append(f"Imputed {len(cat_cols)} categorical cols with mode")

        n_out = 0
        for col in num_cols:
            z    = np.abs((df[col] - df[col].mean()) / (df[col].std() + 1e-9))
            n_out += (z > self.outlier_z).sum()
            df   = df[z <= self.outlier_z]
        steps.append(f"Removed ~{n_out} outlier data-points (z>{self.outlier_z})")

        df.columns = [re.sub(r'[^\w]', '_', c).lower().strip('_') for c in df.columns]
        steps.append("Normalised column names to snake_case")

        rows = "".join(
            f"<tr><td style='padding:7px 14px; color:#48cae4;'>✔</td>"
            f"<td style='padding:7px 14px; color:#c9d1d9; font-family:monospace;'>{s}</td></tr>"
            for s in steps)
        display(HTML(f"""
        <div style='background:#161b22; border:1px solid #30363d; border-radius:10px;
                    padding:16px; margin:10px 0;'>
          <h4 style='color:#48cae4; margin:0 0 10px;'>🧹 Cleaning Steps Applied</h4>
          <table style='border-collapse:collapse;'><tbody>{rows}</tbody></table>
        </div>"""))

        self.state = {'cleaned_shape': df.shape, 'steps': steps}
        self.status(f"Cleaning done — final shape: {df.shape}")
        return df

print(" DataCleaningAgent defined.")
```

```python
class EDAAgent(BaseAgent):

    MAX_CAT_VALS = 15

    def __init__(self):
        super().__init__('EDAAgent', 'eda')

    def run(self, df: pd.DataFrame, context: dict = None) -> pd.DataFrame:
        self.banner(" EDA Agent")
        num_cols = df.select_dtypes(include=np.number).columns.tolist()
        cat_cols = df.select_dtypes(include='object').columns.tolist()

        if num_cols:                self._plot_distributions(df, num_cols)
        if cat_cols:                self._plot_categories(df, cat_cols)
        if len(num_cols) >= 2:      self._plot_pairplot(df, num_cols[:5])
        self._plot_boxplots(df, num_cols[:8])

        self.state = {'num_cols': num_cols, 'cat_cols': cat_cols}
        self.status(f"EDA complete — {len(num_cols)} numeric, {len(cat_cols)} categorical.")
        return df

    def _plot_distributions(self, df, cols):
        n     = min(len(cols), 12)
        ncols = 3
        nrows = (n + ncols - 1) // ncols
        fig, axes = plt.subplots(nrows, ncols, figsize=(15, 4*nrows))
        axes = np.array(axes).flatten()
        for i, col in enumerate(cols[:n]):
            ax = axes[i]
            ax.hist(df[col].dropna(), bins=30,
                    color=PALETTE[i % len(PALETTE)], edgecolor='none', alpha=0.85)
            ax.set_title(col, color='#c9d1d9', fontsize=10)
            ax.axvline(df[col].mean(),   color='#e94560', lw=1.5, linestyle='--', label='mean')
            ax.axvline(df[col].median(), color='#48cae4', lw=1.5, linestyle=':',  label='median')
            ax.legend(fontsize=7)
        for j in range(i+1, len(axes)): axes[j].set_visible(False)
        fig.suptitle('📈 Numerical Distributions', color='#a8dadc', fontsize=15, y=1.01)
        plt.tight_layout(); plt.show()

    def _plot_categories(self, df, cols):
        plot_cols = [c for c in cols if df[c].nunique() <= self.MAX_CAT_VALS][:6]
        if not plot_cols: return
        nrows = (len(plot_cols) + 1) // 2
        fig, axes = plt.subplots(nrows, 2, figsize=(14, 4*nrows))
        axes = np.array(axes).flatten()
        for i, col in enumerate(plot_cols):
            vc = df[col].value_counts().head(self.MAX_CAT_VALS)
            axes[i].barh(vc.index.astype(str), vc.values,
                         color=[PALETTE[j % len(PALETTE)] for j in range(len(vc))])
            axes[i].set_title(col, color='#c9d1d9', fontsize=10)
            axes[i].invert_yaxis()
        for j in range(i+1, len(axes)): axes[j].set_visible(False)
        fig.suptitle(' Categorical Distributions', color='#a8dadc', fontsize=15, y=1.01)
        plt.tight_layout(); plt.show()

    def _plot_pairplot(self, df, cols):
        sample = df[cols].dropna().sample(min(300, len(df)), random_state=42)
        n = len(cols)
        fig, axes = plt.subplots(n, n, figsize=(3*n, 3*n))
        for i, ci in enumerate(cols):
            for j, cj in enumerate(cols):
                ax = axes[i][j]
                if i == j:
                    ax.hist(sample[ci], bins=20, color=PALETTE[i % len(PALETTE)], alpha=0.8)
                else:
                    ax.scatter(sample[cj], sample[ci], alpha=0.3, s=8,
                               color=PALETTE[(i+j) % len(PALETTE)])
                if j == 0:    ax.set_ylabel(ci, fontsize=8)
                if i == n-1:  ax.set_xlabel(cj, fontsize=8)
        fig.suptitle(' Pair Plot', color='#a8dadc', fontsize=14)
        plt.tight_layout(); plt.show()

    def _plot_boxplots(self, df, cols):
        if not cols: return
        fig, ax = plt.subplots(figsize=(max(8, len(cols)*1.2), 5))
        bp = ax.boxplot([df[c].dropna() for c in cols], patch_artist=True,
                        medianprops=dict(color='#e94560', lw=2))
        for patch, color in zip(bp['boxes'], PALETTE):
            patch.set_facecolor(color + '55')
            patch.set_edgecolor(color)
        ax.set_xticklabels(cols, rotation=30, ha='right')
        ax.set_title(' Box Plots — Outlier Overview', color='#a8dadc', fontsize=13)
        plt.tight_layout(); plt.show()

print(" EDAAgent defined.")
```

```python
class AutoMLAgent(BaseAgent):

    CLASSIFIERS = [
        ('RandomForest',     RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)),
        ('XGBoost',          XGBClassifier(random_state=42, eval_metric='logloss', verbosity=0)),
        ('LightGBM',         LGBMClassifier(random_state=42, verbose=-1)),
        ('GradientBoosting', GradientBoostingClassifier(random_state=42)),
        ('LogisticReg',      LogisticRegression(max_iter=1000, random_state=42)),
    ]
    REGRESSORS = [
        ('RandomForest', RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)),
        ('XGBoost',      XGBRegressor(random_state=42, verbosity=0)),
        ('LightGBM',     LGBMRegressor(random_state=42, verbose=-1)),
        ('Ridge',        Ridge()),
    ]

    def __init__(self, target: str, test_size: float = 0.2):
        super().__init__('AutoMLAgent', 'ml')
        self.target    = target
        self.test_size = test_size

    def run(self, df: pd.DataFrame, context: dict = None) -> pd.DataFrame:
        self.banner(" AutoML Agent")

        if self.target not in df.columns:
            self.status(f"Target '{self.target}' not found — skipping ML.", ok=False)
            return df

        X, y = self._prepare(df)
        task = 'classification' if y.nunique() <= 20 and y.dtype != float else 'regression'
        self.status(f"Task: {task.upper()} | unique targets: {y.nunique()}")

        X_tr, X_te, y_tr, y_te = train_test_split(
            X, y, test_size=self.test_size, random_state=42,
            stratify=y if task == 'classification' else None)

        scaler = RobustScaler()
        X_tr   = scaler.fit_transform(X_tr)
        X_te   = scaler.transform(X_te)

        models  = self.CLASSIFIERS if task == 'classification' else self.REGRESSORS
        results = self._train_models(models, X_tr, X_te, y_tr, y_te, task)
        self._plot_leaderboard(results, task)

        best_name, best_score, best_model = results[0]
        self.status(f" Best: {best_name} — score: {best_score:.4f}")
        self._feature_importance(best_model, X.columns.tolist(), best_name)

        if task == 'classification':
            self._plot_confusion(best_model, X_te, y_te, best_name)

        self.state = {
            'task': task, 'best_model': best_name,
            'best_score': best_score, 'results': results
        }
        return df

    def _prepare(self, df):
        df2 = df.copy()
        y   = df2.pop(self.target)
        for col in df2.select_dtypes(include='object').columns:
            df2[col] = LabelEncoder().fit_transform(df2[col].astype(str))
        if y.dtype == object:
            y = pd.Series(LabelEncoder().fit_transform(y.astype(str)))
        return df2, y

    def _train_models(self, models, X_tr, X_te, y_tr, y_te, task):
        results = []
        for name, mdl in models:
            try:
                mdl.fit(X_tr, y_tr)
                if task == 'classification':
                    score = f1_score(y_te, mdl.predict(X_te), average='weighted')
                else:
                    score = r2_score(y_te, mdl.predict(X_te))
                results.append((name, round(score, 4), mdl))
                self.status(f"{name}: {'F1' if task=='classification' else 'R²'} = {score:.4f}")
            except Exception as e:
                self.status(f"{name} failed: {e}", ok=False)
        results.sort(key=lambda x: x[1], reverse=True)
        return results

    def _plot_leaderboard(self, results, task):
        names  = [r[0] for r in results]
        scores = [r[1] for r in results]
        fig, ax = plt.subplots(figsize=(9, 4))
        bars = ax.barh(names, scores, color=PALETTE[:len(names)], edgecolor='none', height=0.55)
        for bar, score in zip(bars, scores):
            ax.text(score + 0.002, bar.get_y() + bar.get_height()/2,
                    f'{score:.4f}', va='center', color='#c9d1d9', fontsize=9)
        metric = 'F1 (weighted)' if task == 'classification' else 'R²'
        ax.set_title(f' Model Leaderboard — {metric}', color='#ffb74d', fontsize=13)
        ax.set_xlabel(metric); ax.invert_yaxis()
        plt.tight_layout(); plt.show()

    def _feature_importance(self, model, feature_names, model_name):
        if not hasattr(model, 'feature_importances_'): return
        fi  = pd.Series(model.feature_importances_, index=feature_names).sort_values(ascending=True)
        top = fi.tail(15)
        fig, ax = plt.subplots(figsize=(9, max(4, len(top)*0.45)))
        top.plot.barh(ax=ax,
                      color=plt.cm.RdYlGn(np.linspace(0.2, 0.9, len(top))),
                      edgecolor='none')
        ax.set_title(f' Feature Importances — {model_name}', color='#ffb74d', fontsize=13)
        plt.tight_layout(); plt.show()

    def _plot_confusion(self, model, X_te, y_te, model_name):
        cm = confusion_matrix(y_te, model.predict(X_te))
        fig, ax = plt.subplots(figsize=(6, 5))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                    linewidths=0.5, cbar=False, annot_kws={'color': 'white', 'size': 11})
        ax.set_title(f' Confusion Matrix — {model_name}', color='#ffb74d', fontsize=12)
        ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')
        plt.tight_layout(); plt.show()

print(" AutoMLAgent defined.")
```

```python
class ReportAgent(BaseAgent):

    def __init__(self):
        super().__init__('ReportAgent', 'reporter')

    def run(self, df: pd.DataFrame, context: dict = None) -> pd.DataFrame:
        self.banner(" Report Generation Agent")
        ctx      = context or {}
        profiler = ctx.get('DataProfilingAgent', {})
        cleaner  = ctx.get('DataCleaningAgent',  {})
        ml       = ctx.get('AutoMLAgent',        {})

        shape   = profiler.get('shape', df.shape)
        c_shape = cleaner.get('cleaned_shape', df.shape)
        task    = ml.get('task',       'N/A')
        best    = ml.get('best_model', 'N/A')
        score   = ml.get('best_score', 0)
        metric  = 'F1 (weighted)' if task == 'classification' else 'R²'

        steps_html = "".join(
            f"<li style='color:#c9d1d9; margin:4px 0; font-family:monospace;'>{s}</li>"
            for s in cleaner.get('steps', []))

        ml_rows = "".join(
            f"<tr><td style='padding:6px 14px; color:#48cae4;'>{n}</td>"
            f"<td style='padding:6px 14px; color:#81c784;'>{s:.4f}</td></tr>"
            for n, s, _ in ml.get('results', []))

        display(HTML(f"""
        <div style='background:linear-gradient(135deg,#0d1117,#161b22);
                    border:1px solid #30363d; border-radius:16px;
                    padding:30px; font-family:monospace; margin:20px 0;'>

          <div style='text-align:center; margin-bottom:25px;'>
            <h1 style='color:#81c784; font-size:2em; margin:0;'>📋 Analysis Report</h1>
            <p style='color:#8b949e; margin:6px 0;'>
              Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
            </p>
          </div>

          <div style='background:#161b22; border-radius:10px; padding:16px; margin:12px 0;
                      border-left:3px solid #e94560;'>
            <h3 style='color:#e94560; margin:0 0 10px;'>🔍 Data Overview</h3>
            <p style='color:#c9d1d9; margin:4px 0;'>
              Raw shape: <b style='color:#48cae4;'>{shape[0]:,} rows × {shape[1]} cols</b>
            </p>
            <p style='color:#c9d1d9; margin:4px 0;'>
              After cleaning: <b style='color:#81c784;'>{c_shape[0]:,} rows × {c_shape[1]} cols</b>
            </p>
            <p style='color:#c9d1d9; margin:4px 0;'>
              Missing values: <b style='color:#ffb74d;'>{profiler.get('missing_total','N/A')}</b>
            </p>
          </div>

          <div style='background:#161b22; border-radius:10px; padding:16px; margin:12px 0;
                      border-left:3px solid #48cae4;'>
            <h3 style='color:#48cae4; margin:0 0 10px;'>🧹 Cleaning Steps</h3>
            <ul style='margin:0; padding-left:20px;'>{steps_html}</ul>
          </div>

          <div style='background:#161b22; border-radius:10px; padding:16px; margin:12px 0;
                      border-left:3px solid #ffb74d;'>
            <h3 style='color:#ffb74d; margin:0 0 10px;'> ML Results</h3>
            <p style='color:#c9d1d9; margin:4px 0;'>
              Task: <b style='color:#ce93d8;'>{task.upper()}</b>
            </p>
            <p style='color:#c9d1d9; margin:4px 0;'>
              Best Model: <b style='color:#e94560;'>{best}</b>
              | {metric}: <b style='color:#81c784;'>{score:.4f}</b>
            </p>
            <table style='border-collapse:collapse; margin-top:10px;'>
              <thead><tr style='background:#21262d;'>
                <th style='padding:6px 14px; color:#c9d1d9;'>Model</th>
                <th style='padding:6px 14px; color:#c9d1d9;'>{metric}</th>
              </tr></thead><tbody>{ml_rows}</tbody>
            </table>
          </div>

          <div style='text-align:center; margin-top:20px; padding-top:15px;
                      border-top:1px solid #30363d;'>
            <p style='color:#8b949e; font-size:0.85em;'>
               Multi-Agent Data Analyst System
            </p>
          </div>
        </div>"""))

        self.status("Report generated successfully.")
        self.state = {'report_generated': True}
        return df

print(" ReportAgent defined.")
```
